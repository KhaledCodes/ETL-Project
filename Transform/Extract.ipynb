```python
# Extracting all CSVs to Dataframes
Canada=pd.read_csv('CAvideos.csv')
```

```python
# Extracting all JSONs into Dataframes
CaJSON=pd.read_json('CA_category_id.json')
```

```python
# Creating a new column for the category name
Canada['Group']=''
```

```python
# Assigning the category name to the 'Group' column by referencing the JSON file (which has the category name for each entry)
for row in range(len(Canada)):
    for i in range(len(CaJSON)):
        if (str(Canada.loc[row,'category_id'])==(CaJSON)['items'][i]['id']):
            Canada.set_value(row,('Group'),(CaJSON['items'][i]['snippet']['title']))
```

```python
# Saving the new dataframe as a csv
Canada.to_csv(f"Canada_DF.csv")
```

```python
# Looping through the top 100 videos and getting the top comment for each video and appending it to a list
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
comments=[]
titles=[]
for m in range(100):
    url=f"https://www.youtube.com/watch?v={Canada.loc[m,'video_id']}"
    driver = webdriver.Chrome("chromedriver.exe")
    driver.get(url)
    try:
        time.sleep(2)
        driver.find_element_by_tag_name('body').send_keys(Keys.PAGE_DOWN)
        time.sleep(2)
        html = driver.page_source
        soup = bs(html, 'html.parser')
        comment = soup.find_all('yt-formatted-string', class_="style-scope ytd-comment-renderer", id="content-text",slot='content')
        title=soup.find_all
        comments.append(comment[0].text)
        titles.append(soup.title.text)
    except:
        pass
```

```python
# Transforming the list into a DataFrame
titlecomment=pd.DataFrame(list(zip(titles,comments)),columns=['Title','Comments'])
```

```python
# Transforming the DataFrame into a CSV
titlecomment.to_csv("Title and Comments.csv")
```
